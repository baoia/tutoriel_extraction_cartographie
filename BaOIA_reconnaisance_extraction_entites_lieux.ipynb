# **Reconnaissance des Entités nommées (REN)**:
Script de reconnaissance et extraction et entités nommées de textes. Entités reconnues: LOC (lieux géographiques ou politiques)

Bibliothèque python utilisée: SpaCy.

Documents d'entrée: fichiers .txt contenant les textes sur lesquels effectuer la reconnaissane des entitiés nommées.

Fichiers de sortie : 1 fichier .txt contenant une liste des entités nommées (lieux) pour chaque texte d'origine

```
#@markdown # Reconnaissance et téléchargement des entités de lieux

chemin_vers_le_dossier_de_travail = '/content/drive/My Drive/tutoriel_extraction_cartographie-main/'#@param {type:"string"}
os.chdir(chemin_vers_le_dossier_de_travail)
nlp.max_length = 150000000
chemin_vers_le_dossier_de_fichiers_txt = "/content/drive/My Drive/tutoriel_extraction_cartographie-main/fichiers textes guides gallica/"#@param {type:"string"}
nom_du_dossier_des_entites_de_lieux = "entites lieux guides" #@param {type:"string"}
if not os.path.exists(nom_du_dossier_des_entites_de_lieux):
  os.mkdir(nom_du_dossier_des_entites_de_lieux)
os.chdir(nom_du_dossier_des_entites_de_lieux)
nom_glob = ''.join(chemin_vers_le_dossier_de_fichiers_txt+"*.txt")
for element in glob.glob(nom_glob):
  with open (element, "r") as myfile:
    data=myfile.read()
    doc = nlp(data)
    nom_fichier = os.path.basename(element)
    nouveau_nom = ''.join("Lieux_" + nom_fichier)
    with open (nouveau_nom, 'w') as my_locs:
      for ent in doc.ents:
        if ent.label_ == 'LOC':
          entites_loc = "".join([str(ent.text),"\n"])
          my_locs.write(entites_loc)
```
